## Value Iteration

```python
# Value Iteration
import pandas as pd
P, R = hiive.mdptoolbox.example.forest(1000)
gamma = 0.99
vi = hiive.mdptoolbox.mdp.ValueIteration(P, R, gamma)
to_run = vi.run()
n_iter = vi.iter
state_run = pd.DataFrame(to_run)
```

```python
state_run
```

```python
algo = 'Forest - Value Iterations:'
fig, ax = plt.subplots()
ax.plot(state_run['Iteration'] , state_run['Max V'], "r-")
ax.set_xlabel('Iterations')
ax.set_ylabel('Max V')
ax.set_title(f'{algo} Convergence Criteria')
# ax.legend()
plt.show(fig)

fig, ax = plt.subplots()
ax.plot(state_run['Iteration'] , state_run['Mean V'], "r-")
ax.set_xlabel('Iterations')
ax.set_ylabel('Avg V')
ax.set_title(f'{algo} Convergence Criteria')
# ax.legend()
plt.show(fig)

fig, ax = plt.subplots()
ax.plot(state_run['Iteration'] , state_run['Error'], "r-")
ax.set_xlabel('Iterations')
ax.set_ylabel('Errors')
ax.set_title(f'{algo} Convergence Criteria')
# ax.legend()
plt.show(fig)

fig, ax = plt.subplots()
ax.plot(state_run['Iteration'] , state_run['Time'], "b-")
ax.set_xlabel('Iterations')
ax.set_ylabel('Computation Time')
ax.set_title(f'{algo} Time and Iterations to Converge')
# ax.legend()
plt.show(fig)
```

```python
# Value Iteration!!
gamma = 0.99 
to_loop = np.arange(0.1, 1 , 0.01)
# P, R = hiive.mdptoolbox.example.forest(1000)
# 4x4 !! 
vi_util_opt_pol = []
vi_num_iterations = []
vi_time_taken_list = []
vi_rewards_list = []
for i in to_loop:
    print("Discount Factor(gamma):", i)
    vi = hiive.mdptoolbox.mdp.ValueIteration(P, R, i)
    vi.run()
    vi_rewards_list.append([np.max(vi.V), np.mean(vi.V)])
    vi_util_opt_pol.append(vi.policy)
    vi_num_iterations.append(vi.iter)
    vi_time_taken_list.append(vi.time)
    print("Utility of optimal policy:" ,(np.max(vi.V), np.mean(vi.V)), "Time Taken:", vi.time, 'Num Iterations:', vi.iter)
```

```python
import matplotlib.pyplot as plt
algo = 'Forest - Value Iterations:'
fig, ax = plt.subplots()
ax.plot(list(to_loop) , [i[0] for i in vi_rewards_list], "g-")
ax.set_xlabel('Discount Factor(Gamma)')
ax.set_ylabel('Utilities (Rewards)')
ax.set_title(f'{algo} Utilities vs Discount Factor')
# ax.legend()
plt.show(fig)

fig, ax = plt.subplots()
ax.plot(list(to_loop) , [i[1] for i in vi_rewards_list], "g-")
ax.set_xlabel('Discount Factor(Gamma)')
ax.set_ylabel('Utilities (Rewards)')
ax.set_title(f'{algo} Utilities (Avg) vs Discount Factor')
# ax.legend()
plt.show(fig)


fig, ax = plt.subplots()
ax.plot(list(to_loop) , vi_num_iterations, "b-")
ax.set_xlabel('Discount Factor(Gamma)')
ax.set_ylabel('Number of Iterations')
ax.set_title(f'{algo} Number of Iterations vs Discount Factor')
# ax.legend()
plt.show(fig)

fig, ax = plt.subplots()
ax.plot(list(to_loop) , vi_time_taken_list, "r-")
ax.set_xlabel('Discount Factor(Gamma)')
ax.set_ylabel('Time Taken (Seconds)')
ax.set_title(f'{algo} Time Taken (Seconds) vs Discount Factor')
# ax.legend()
plt.show(fig)
```

```python
# Q Learning!
gamma = 0.9
epsilon = 0.9
to_loop = np.arange(0.1, 1 , 0.1)
P, R = hiive.mdptoolbox.example.forest(1000)
# 4x4 !! 
util_opt_pol = []
error_list = []
time_taken_list = []
rewards_list = []
for i in to_loop:
    print("Epsilon:", i)
    ql = hiive.mdptoolbox.mdp.QLearning(P, R, gamma, epsilon=i, epsilon_decay=0.999, n_iter=100000)
    ql.run()
    rewards_list.append([np.max(ql.V), np.mean(ql.V)])
    util_opt_pol.append(ql.policy)
    time_taken_list.append(ql.time)
    error_list.append(ql.error_mean)
    print("Utility of optimal policy:" ,(np.max(ql.V), np.mean(ql.V)), "Time Taken:", ql.time)
```

```python
algo = 'QLearning:'
fig, ax = plt.subplots()
ax.plot(list(to_loop) , [i[0] for i in rewards_list], "g-")
ax.set_xlabel('Epsilon')
ax.set_ylabel('Utilities (Rewards)')
ax.set_title(f'{algo} Utilities vs different Epsilon')
# ax.legend()
plt.show(fig)

fig, ax = plt.subplots()
ax.plot(list(to_loop) , [i[1] for i in rewards_list], "g-")
ax.set_xlabel('Epsilon')
ax.set_ylabel('Utilities (Rewards)')
ax.set_title(f'{algo} Utilities (AVG) vs different Epsilon')
# ax.legend()
plt.show(fig)

fig, ax = plt.subplots()
ax.plot(list(to_loop) , error_list, "b-")
ax.set_xlabel('Epsilon')
ax.set_ylabel('Mean Error')
ax.set_title(f'{algo} Mean Errors vs different Epsilon')
# ax.legend()
plt.show(fig)

fig, ax = plt.subplots()
ax.plot(list(to_loop) , time_taken_list, "r-")
ax.set_xlabel('Epsilon')
ax.set_ylabel('Time Taken (Seconds)')
ax.set_title(f'{algo} Time Taken (Seconds) vs different Epsilon')
# ax.legend()
plt.show(fig)
```

```python
action_mapping_n = {0:'W', 1:'C'}
new_cnt = 0
def cumulative_cut_plot_policy(policy):
    new_cnt = 0
    new_text_sum = ""
    for i in policy:
        if new_cnt % 100 == 0:
            new_text_sum += "\n"
        new_text_sum += action_mapping_n[i]
        new_cnt +=1
    return new_text_sum
```

```python
print(cumulative_cut_plot_policy(ql.policy))
```
